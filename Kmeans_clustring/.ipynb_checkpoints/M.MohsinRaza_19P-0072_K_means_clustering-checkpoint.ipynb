```python
class clustering:
    def __init__(self):
        
        self.centroids = X[centroidIndex]
        print('centroids:',self.centroids)
        self.X_ = np.delete(X,centroidIndex, axis = 0)
    
            
    def euclidian_distance(self,query,X):
        difference = np.array(X) - np.array(query)
        sqrd_diff = np.square(difference)
        sum_sqrd_diff = np.sum(sqrd_diff, axis = 1)
        distance = np.sqrt(sum_sqrd_diff)
        return distance
        
        
    def assign_clusters(self):
        self.clusters = [[],[],[],[]]
        
        for i in range (4):
            self.clusters[i].append(self.centroids[i])
            
        for x in self.X_:
            id = np.argmin(self.euclidian_distance(x,self.centroids))
            self.clusters[id].append(x)
        return self.clusters
            
    def assign_centroids(self):
        for i in range(4):
            self.centroids[i] = np.mean(self.clusters[i],axis = 0)
        return self.centroids
    
    def get_centroids(self):
        return self.centroids
    
    def get_clusters(self):
        return self.clusters
            
    def print_val(self):
        print('centroids:',self.centroids)
        print('centroids:',self.clusters)
        
```

```python
c=clustering()
```

```python
clusters=c.assign_clusters()
```

```python
old_cen,old_clusters=0,0
for i in range(100):
    
    centroids=c.get_centroids().copy()
    newCentroids=c.assign_centroids()
    clusters=c.get_clusters().copy()
    newClusters=c.assign_clusters()
    
    print("Centroids:",centroids)
    print('\n')
    print("Clusters:",len(clusters[0]),'\n')
    print("Clusters:",len(clusters[1]),'\n')
    print("Clusters:",len(clusters[2]),'\n')
    print("Clusters:",len(clusters[3]),'\n')
    
    print("newCentroids:",newCentroids)
    print('\n')
    print("newClusters:",len(newClusters[0]),'\n')
    print("newClusters:",len(newClusters[1]),'\n')
    print("newClusters:",len(newClusters[2]),'\n')
    print("newClusters:",len(newClusters[3]),'\n')
    
    #---- can u as value of K------#
    clusters_result=[[],[],[],[]]
    centroids_result=[[],[],[],[]]
    c1=[]
    c2=[]
    for x in range(4):
        #--- np.array_equal-> True if two arrays have the same shape and elements, False otherwise.---#
        clusters_result[x]=np.array_equal(np.array(clusters[x]),np.array(newClusters[x]))
        c1.append(np.array(clusters_result[x]).all())
        centroids_result[x]=np.array_equal(np.array(centroids[x]),np.array(newCentroids[x]))
        c2.append(np.array(centroids_result[x]).all())
        
        
    if np.logical_or(np.array(c1).all(),np.array(c2).all()):
#     if np.logical_or(np.all(centroids ==newCentroids ) , np.all(clusters == newClusters) ):
            print('------No change required-------','\n')
            print('Centroids: ',centroids,'\n')
            print('Clusters: ',clusters)
            break
    
```
