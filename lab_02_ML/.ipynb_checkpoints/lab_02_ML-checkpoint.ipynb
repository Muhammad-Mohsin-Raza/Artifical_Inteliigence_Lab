```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv('fruit_data_with_colors.txt', delimiter='\t')
df
```

```python
df.head()
```

```python
# query_instance=np.array([180,8.0,6.8,0.59])
# mass	width	height	color_score
# w,x,y,z
#mass-x

X=df[['mass','width','height','color_score']]
Y=df['fruit_label']

# num_arr=np.array(df[['mass','width','height','color_score']])
# df.iloc[:50]

trainX=np.array(X[:50])
trainY=np.array(Y[:50])

testX=np.array(X[50:])
testY=np.array(Y[50:])

trainX[0]
```

```python
class KNN:
    def __init__(self,k=1):
        self.k = k
        
    def euclidian_dist(self,target_query,X):
        
        sq_point_dist=np.square((np.array(X)-np.array(target_query)))
        sum_of_sq=np.sum(sq_point_dist, axis=1)
        distance=np.sqrt(sum_of_sq)
        return distance
    
    def nearest_neighbours(self,distance):
        
        return np.argsort(distance)[:self.k]
                
    def predict(self,query,trainX,trainY):
        ed=self.euclidian_dist(query,trainX)
        nn=self.nearest_neighbours(ed)
        labels_nn = list(trainY[nn])
        return max(labels_nn, key = labels_nn.count) 
        
```

```python
classifier = KNN(3)
predictions = [classifier.predict(x,trainX,trainY) for x in testX]
predictions
#---- 9 instances so 9 lables for each instance it will pick K distance lables of min dis---#
```

```python
#confusion matrix , performance_accuracy skitlearn
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predictions,testY)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(testY,predictions)
```

```python
for i in range (1,6):
    print('\n')
    print('K value = ',i)
    classifier = KNN(i)
    predictions = [classifier.predict(x,trainX,trainY) for x in testX]
    print(predictions)
    
    print('Accuracy Score',accuracy_score(predictions,testY))
    print('Confusion Matrix',confusion_matrix(testY,predictions))
```
